##  1. The Problem Before Transformers
- Traditional models like RNNs(Recurrent Neural Network) and LSTMs processed text word by word.
- **Issues:** Long-term dependencies, memory problems, and slow training
- These models couldn’t handle **parallelism** well, which made them inefficient for large-scale data

*A model was required that could handle entire sequences at once and remember context better.*

##  2. Enter Transformers: The Breakthrough
Introduced in the paper *“Attention is All You Need”* (Vaswani et al., 2017).

- Process entire input **in parallel**
- Use **Self-Attention** to relate every word to every other word
- Eliminate the need for recurrence (like RNNs)

**Key Innovation: Self-Attention**
> "She poured water in the glass because it was *thirsty*."
→ A Transformer uses attention to understand that *"it"* refers to the *glass*, not *she*.

##  3. Understanding the Transformer Architecture

Here's the full architecture — split into two main components: **Encoder** and **Decoder**.
transformer.png
###  Encoder (Left Side)
- **Embeddings + Positional Encoding**: Words are converted into vectors, and positional info is added to keep track of word order.
- **Multi-Head Self-Attention**: Each word attends to (reads) all other words in the sentence. Helps understand relationships like *"it" ->"glass"*.
- **Feedforward Layers (MLPs)**: Processes each word’s vector further.
- input data from attention -> 1st layer of mlp -> non linearity is applied (using reLu function) -> 2nd layer -> final activation
- **Add & Norm**: Shortcut connections that make learning stable.

###  Decoder (Right Side)
- **Masked Self-Attention**: Prevents the model from looking at future words while generating.
- **Encoder-Decoder Attention**: Connects output words with input context (e.g., translating a sentence).
- **Feedforward + Add & Norm**: Just like the encoder.

###  Output
- Passed through a **Linear layer** and **Softmax** to generate word predictions.

This is where attention replaces recurrence — and makes training fast and powerful.


##  4. Transformer Architecture (Simplified)
- **Input Embedding**: Converts tokens to vectors
- **Positional Encoding**: Adds info about word order
- **Multi-Head Attention**: Learns different kinds of relationships
- **Feedforward Layers**: Refines the understanding
- **Residual + Layer Norm**: Stabilizes training



##  5. Transformers Are Still Neural Networks
- They use **layers, weights, activations**, and **backpropagation**
- Self-Attention is just a specialized neural layer
